```python
## To Perform Merge Function
```

```python
data1=pd.DataFrame({'key1':[3,46,17,98,23],
                   'key2':[45,12,17,81,23],
                   'key3':[37,91,26,65,34]})
```

```python
data2=pd.DataFrame({'key1':[3,46,17,89,23],
                   'key4':[45,21,17,18,23],
                   'key5':[73,91,62,65,34]})
```

```python
data1
```

```python
data2
```

```python
pd.merge(data1,data2,how="inner")
```

```python
pd.merge(data1,data2,how="left")
```

```python
pd.merge(data1,data2,how="outer")
```

```python
pd.merge(data1,data2,how="cross")
```
